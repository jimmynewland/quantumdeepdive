# Alice generates a secret key.

Define a string called secretKeyAlice of length n that is composed of a random sequence of '0' and '1' characters.

```python
import random

MESSAGE_LENGTH = 10 # length of the text message we want to send (We can always add more later.)

n = 8  * MESSAGE_LENGTH # number of qubits  (We multiply by 8 because each character is represented by 8 bits in ASCII.)

secretKeyAlice = ""

# TODO: Put your code here.
for i in range(n):
  secretKeyAlice += random.choice(['0','1'])
# -------------------------

print("secretKeyAlice = " + secretKeyAlice) # print out Alice's secret key

# Sample Output:
# secretKeyAlice    = 01110111001011001100001011000100001101001110111000011100101111101000000010101110
```

# Bob copies Alice's secret key

Define a string called secretKeyBob that is an exact copy of secretKeyAlice.

```python
secretKeyBob = ""

# TODO: Put your code here.
for i in range(n):
  secretKeyBob += secretKeyAlice[i]
# -------------------------

print("secretKeyBob = " + secretKeyBob) # print out Bob's secret key
```

## Compare Alice and Bob's secret keys.
Run this cell.  You should find that Alice and Bob's sifted keys match 100%.

```python
numMatch = 0

if len(secretKeyAlice) != len(secretKeyBob):
    print("Secret keys are different lengths!")
else:
    for i in range(len(secretKeyAlice)):
        if secretKeyAlice[i] == secretKeyBob[i]:
           numMatch += 1
    matchPercent = numMatch / len(secretKeyAlice) * 100

print(str(matchPercent) + "% match")

# Sample Output:
# 100.0% match
```
